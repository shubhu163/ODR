```python
from __future__ import print_function
import pandas as pd
import shutil
import os
import sys

labels = pd.read_csv('data/Filenames.csv')

train_dir = 'data/preprocessed_images'
DR = 'data/dataset'
if not os.path.exists(DR):
    os.mkdir(DR)

for filename, class_name in labels.values:
    if not os.path.exists(DR + str(class_name)):
        os.mkdir(DR + str(class_name))
    src_path = train_dir + '/'+ filename
    dst_path = DR + str(class_name) + '/' + filename 
    try:
        shutil.copy(src_path, dst_path)
        print("sucessful")
    except IOError as e:
        print('Unable to copy file {} to {}'
              .format(src_path, dst_path))
    except:
        print('When try copy file {} to {}, unexpected error: {}'
              .format(src_path, dst_path, sys.exc_info()))
```

```python
import numpy as np
import cv2, glob
import numpy as np
import os
import pandas as pd
from pathlib import Path
```

```python
#COLOR CLAHE FUNC
def enhance(input_path,output_path,clip_limit=3):
    image = cv2.imread(input_path, 1)
    # convert image to LAB color model
    image_lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

    # split the image into L, A, and B channels
    l_channel, a_channel, b_channel = cv2.split(image_lab)

    # apply CLAHE to lightness channel
    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=(8, 8))
    cl = clahe.apply(l_channel)

    # merge the CLAHE enhanced L channel with the original A and B channel
    merged_channels = cv2.merge((cl, a_channel, b_channel))

    # convert iamge from LAB color model back to RGB color model
    final_image = cv2.cvtColor(merged_channels, cv2.COLOR_LAB2BGR)
    cv2.imwrite(output_path,final_image)
#     return final_image
```
